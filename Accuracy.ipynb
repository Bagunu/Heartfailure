```python
dataset = pd.read_csv("heart_failure_clinical_records_dataset.csv")
# df2 = pd.read_csv("./Final Project/datasets_heart.csv")
# Drop the null columns where all values are null
dataset = dataset.dropna(axis='columns', how='all')
# Drop the null rows
dataset = dataset.dropna()
dataset.head()
```

```python
dataset = dataset.rename(columns = {'age': 'age', 'sex': 'sex', 'cp': 'chest pain', 'trestbps': 'blood pressure', 'chol': 'cholestrol level', 'thalach': 'max heart rate', })
dataset.head()
```

```python
# X=combined_health_df[combined_health_df.columns[combined_health_df.columns!='target']]
X= dataset[dataset.columns[dataset.columns!='DEATH_EVENT']]
X.head()
```

```python
Y=dataset['DEATH_EVENT']
Y.head()
```

```python
X.corrwith(Y)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y,test_y=train_test_split(X,Y,random_state=0,test_size=0.2)
print(train_y.shape, test_y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
best_estimator=0
best_depth=0
acc=0
for i in range(1,10):
    for j in range(1,10):
        classifier=RandomForestClassifier(n_estimators=i,max_depth=j,random_state=0,class_weight={0:sum(train_y==0),1:sum(train_y==1)})
        classifier.fit(train_x,train_y)
        if(classifier.score(test_x,test_y) > acc):
            best_estimator=i
            best_depth=j
            acc=classifier.score(test_x,test_y)
print('Best Estimator: ',best_estimator,'\nBest Estimator: ',best_depth,'\nAccuracy: ',acc)
```

```python
best_classifier=RandomForestClassifier(n_estimators=8,max_depth=5,random_state=0,class_weight={0:sum(train_y==0),1:sum(train_y==1)})
best_classifier.fit(train_x,train_y)
print('Train - Accuracy: ',round(best_classifier.score(train_x,train_y)*100,2),'%')
print('Test - Accuracy: ',round(best_classifier.score(test_x,test_y)*100,2),'%')
```

```python
from sklearn.metrics import confusion_matrix
cfmt=confusion_matrix(test_y,best_classifier.predict(test_x))
print(cfmt)
```
